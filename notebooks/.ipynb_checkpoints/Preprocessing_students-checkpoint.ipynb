```python
from tabula import read_pdf
import pandas as pd

df = read_pdf("../data/raw/students/prishtine_numri_i_pergjithshem_bachelor.pdf")
```

```python
# TODO(Add Master's values)
df_up = df.iloc[:,-3:]
df_up.columns = ["Femra", "Meshkuj", "Gjithsej"]
df_up = df_up.iloc[16]
# df_up["Emri"] = "Universiteti i Prishtinës"

df_prishtine_master = pd.read_csv(
    "../data/raw/students/prishtine_master.csv",
    encoding="iso-8859-1",
    sep=";",
    header=1,
    na_values="-",
)

df_all = pd.DataFrame([[sum(x) for x in zip(list(df_up.astype(float).values), [
    df_prishtine_master.iloc[-1,2::2].sum(),
    df_prishtine_master.iloc[-1,3::2].sum(),
    df_prishtine_master.iloc[-1,2::2].sum()+df_prishtine_master.iloc[-1,3::2].sum(),
])]+["Universiteti i Prishtinës"]], columns=["Femra", "Meshkuj", "Gjithsej", "Emri"])
```

```python
df_ferizaj_bachelor = pd.read_csv(
    "../data/raw/students/ferizaj_bachelor.csv",
    encoding="iso-8859-1",
    sep=";",
    header=1,
    na_values="-",
)
df_ferizaj_master = pd.read_csv(
    "../data/raw/students/ferizaj_bachelor.csv",
    encoding="iso-8859-1",
    sep=";",
    header=1,
    na_values="-",
)

df_all = df_all.append(pd.DataFrame([[
    df_ferizaj_bachelor.iloc[-1,2::3].sum()+df_ferizaj_master.iloc[-1,2::3].sum(),
    df_ferizaj_bachelor.iloc[-1,3::3].sum()+df_ferizaj_master.iloc[-1,3::3].sum(),
    df_ferizaj_bachelor.iloc[-1,4::3].sum()+df_ferizaj_master.iloc[-1,4::3].sum(),
    "Universiteti i Shkencave të Aplikuara Ferizaj"
]],columns=df_all.columns))
```

```python
df_gjakove_bachelor = pd.read_csv(
    "../data/raw/students/gjakove_bachelor.csv",
    encoding="iso-8859-1",
    sep=";",
    header=1,
    na_values="-",
)

df_all = df_all.append(pd.DataFrame([[
    df_gjakove_bachelor.iloc[-1,2::3].sum(),
    df_gjakove_bachelor.iloc[-1,3::3].sum(),
    df_gjakove_bachelor.iloc[-1,4::3].sum(),
    "Universiteti i Gjakovës Fehmi Agani"
]],columns=df_all.columns))
```

```python
df_gjilan_bachelor = pd.read_csv(
    "../data/raw/students/gjilan_bachelor.csv",
    encoding="iso-8859-1",
    sep=";",
    header=1,
    na_values="-",
)

df_all = df_all.append(pd.DataFrame([[
    df_gjilan_bachelor.iloc[-1,3::3].sum(),
    df_gjilan_bachelor.iloc[-1,4::3].sum(),
    df_gjilan_bachelor.iloc[-1,2::3].sum(),
    "Universiteti Kadri Zeka Gjilan"
]],columns=df_all.columns))
```

```python
df_islame_bachelor = pd.read_csv(
    "../data/raw/students/islame_bachelor.csv",
    encoding="iso-8859-1",
    sep=";",
    header=1,
    na_values="-",
)

df_islame_master = pd.read_csv(
    "../data/raw/students/islame_master.csv",
    encoding="iso-8859-1",
    sep=";",
    header=1,
    na_values="-",
)

df_all = df_all.append(pd.DataFrame([[
    df_islame_bachelor.iloc[-1,2::3].sum()+df_islame_master.iloc[-1,2::3].sum(),
    df_islame_bachelor.iloc[-1,3::3].sum()+df_islame_master.iloc[-1,3::3].sum(),
    df_islame_bachelor.iloc[-1,4::3].sum()+df_islame_master.iloc[-1,4::3].sum(),
    "Fakulteti i Studimeve Islame"
]],columns=df_all.columns))
```

```python
df_kolegjet_bachelor = pd.read_csv(
    "../data/raw/students/kolegjet_bachelor.csv",
    encoding="iso-8859-1",
    sep=";",
    header=1,
    na_values=["-", ":"],
)

df_kolegjet_master = pd.read_csv(
    "../data/raw/students/kolegjet_master.csv",
    encoding="iso-8859-1",
    sep=";",
    header=1,
    na_values=["-", ":"],
)

df_all = df_all.append(pd.DataFrame([[
    df_kolegjet_bachelor.iloc[-1,2::2].sum()+df_kolegjet_master.iloc[-1,2::2].sum(),
    df_kolegjet_bachelor.iloc[-1,3::2].sum()+df_kolegjet_master.iloc[-1,3::2].sum()-df_kolegjet_bachelor.iloc[-1,2::2].sum()-df_kolegjet_master.iloc[-1,2::2].sum(),
    df_kolegjet_bachelor.iloc[-1,3::2].sum()+df_kolegjet_master.iloc[-1,3::2].sum(),
    "Kolegjet"
]],columns=df_all.columns))

df_all = df_all.append(pd.DataFrame([[
    df_kolegjet_bachelor.iloc[0,2::2].sum()+df_kolegjet_master.iloc[0,2::2].sum(),
    df_kolegjet_bachelor.iloc[0,3::2].sum()+df_kolegjet_master.iloc[0,3::2].sum()-df_kolegjet_bachelor.iloc[0,2::2].sum()-df_kolegjet_master.iloc[0,2::2].sum(),
    df_kolegjet_bachelor.iloc[0,3::2].sum()+df_kolegjet_master.iloc[0,3::2].sum(),
    "Kolegji AAB"
]],columns=df_all.columns))

df_all = df_all.append(pd.DataFrame([[
    df_kolegjet_bachelor.iloc[18,2::2].sum()+df_kolegjet_master.iloc[18,2::2].sum(),
    df_kolegjet_bachelor.iloc[18,3::2].sum()+df_kolegjet_master.iloc[18,3::2].sum()-df_kolegjet_bachelor.iloc[18,2::2].sum()-df_kolegjet_master.iloc[18,2::2].sum(),
    df_kolegjet_bachelor.iloc[18,3::2].sum()+df_kolegjet_master.iloc[18,3::2].sum(),
    "Kolegji UBT"
]],columns=df_all.columns))

df_all = df_all.append(pd.DataFrame([[
    df_kolegjet_bachelor.iloc[4,2::2].sum()+df_kolegjet_master.iloc[4,2::2].sum(),
    df_kolegjet_bachelor.iloc[4,3::2].sum()+df_kolegjet_master.iloc[4,3::2].sum()-df_kolegjet_bachelor.iloc[4,2::2].sum()-df_kolegjet_master.iloc[4,2::2].sum(),
    df_kolegjet_bachelor.iloc[4,3::2].sum()+df_kolegjet_master.iloc[4,3::2].sum(),
    "Kolegji FAMA"
]],columns=df_all.columns))
```
